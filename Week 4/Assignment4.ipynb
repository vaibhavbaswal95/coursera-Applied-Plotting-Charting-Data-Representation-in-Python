```python
import pandas as pd
f = ['SP1 01-02.csv', 'SP1 02-03.csv', 'SP1 03-04.csv', 'SP1 04-05.csv', 'SP1 05-06.csv', 'SP1 06-07.csv', 'SP1 07-08.csv', 'SP1 08-09.csv', 'SP1 09-10.csv', 'SP1 10-11.csv', 'SP1 11-12.csv', 'SP1 12-13.csv', 'SP1 13-14.csv', 'SP1 14-15.csv', 'SP1 15-16.csv', 'SP1 16-17.csv', 'SP1 97-98.csv', 'SP1 98-99.csv', 'SP1 99-00.csv', 'SP1 00-01.csv']
f
```

```python
data = pd.read_csv('SP1 01-02.csv')
data['season'] = str('SP1 01-02.csv')[4:9]
data
```

```python
for i in range(1,len(f)):
    print(f[i])
    df = pd.read_csv(f[i], error_bad_lines=False)
    df['season'] = str(f[i])[4:9]
    data = pd.concat([data,df])
```

```python
list(data.columns)
```

```python
cols = ['AwayTeam','HomeTeam','FTR','Date','season']
df = data[cols]
```

```python
df['Month-Date'], df['Year'] = zip(*df['Date'].apply(lambda x: (str(x)[:5], str(x)[6:])))
```

```python
df
```

```python
rma = df[((df['AwayTeam'] == 'Real Madrid') & (df['FTR'] == 'A')) | ((df['HomeTeam'] == 'Real Madrid') & (df['FTR'] == 'H')) ].sort(columns = ['Year'])
```

```python
fcb = df[((df['AwayTeam'] == 'Barcelona') & (df['FTR'] == 'A')) | ((df['HomeTeam'] == 'Barcelona') & (df['FTR'] == 'H')) ].sort(columns = ['Year'])
```

```python
atm = df[((df['AwayTeam'] == 'Ath Madrid') & (df['FTR'] == 'A')) | ((df['HomeTeam'] == 'Ath Madrid') & (df['FTR'] == 'H')) ].sort(columns = ['Year'])
```

```python
vil = df[((df['AwayTeam'] == 'Villareal') & (df['FTR'] == 'A')) | ((df['HomeTeam'] == 'Villareal') & (df['FTR'] == 'H')) | ((df['AwayTeam'] == 'Villarreal') & (df['FTR'] == 'A')) | ((df['HomeTeam'] == 'Villarreal') & (df['FTR'] == 'H')) ].sort(columns = ['Year'])
```

```python
vil_w = vil[['FTR','season']].groupby(['season']).agg(['count'])
rma_w = rma[['FTR','season']].groupby(['season']).agg(['count'])
fcb_w = fcb[['FTR','season']].groupby(['season']).agg(['count'])
atm_w = atm[['FTR','season']].groupby(['season']).agg(['count'])
```

```python
rma_w.reset_index(inplace= True)
fcb_w.reset_index(inplace= True)
atm_w.reset_index(inplace= True)
vil_w.reset_index(inplace= True)
```

```python
rma_w
```

```python
rma_w['20s'] = rma_w[(rma_w['season'] != '97-98') & (rma_w['season'] != '98-99') & (rma_w['season'] != '99-00')]['season'].apply(lambda x: '20'+x)
rma_w['19s'] = rma_w[(rma_w['season'] == '97-98') | (rma_w['season'] == '98-99') | (rma_w['season'] == '99-00')]['season'].apply(lambda x: '19'+x)
rma_w
```

```python
fcb_w['20s'] = fcb_w[(fcb_w['season'] != '97-98') & (fcb_w['season'] != '98-99') & (fcb_w['season'] != '99-00')]['season'].apply(lambda x: '20'+x)
fcb_w['19s'] = fcb_w[(fcb_w['season'] == '97-98') | (fcb_w['season'] == '98-99') | (fcb_w['season'] == '99-00')]['season'].apply(lambda x: '19'+x)
atm_w['20s'] = atm_w[(atm_w['season'] != '97-98') & (atm_w['season'] != '98-99') & (atm_w['season'] != '99-00')]['season'].apply(lambda x: '20'+x)
atm_w['19s'] = atm_w[(atm_w['season'] == '97-98') | (atm_w['season'] == '98-99') | (atm_w['season'] == '99-00')]['season'].apply(lambda x: '19'+x)
vil_w['20s'] = vil_w[(vil_w['season'] != '97-98') & (vil_w['season'] != '98-99') & (vil_w['season'] != '99-00')]['season'].apply(lambda x: '20'+x)
vil_w['19s'] = vil_w[(vil_w['season'] == '97-98') | (vil_w['season'] == '98-99') | (vil_w['season'] == '99-00')]['season'].apply(lambda x: '19'+x)
```

```python
rma_w['full_season'] = rma_w['20s'].fillna(rma_w['19s'])
vil_w['full_season'] = vil_w['20s'].fillna(vil_w['19s'])
fcb_w['full_season'] = fcb_w['20s'].fillna(fcb_w['19s'])
atm_w['full_season'] = atm_w['20s'].fillna(atm_w['19s'])
```

```python
season = list(rma_w['full_season'])
season.sort()
season
```

```python
fcb_w.sort('full_season',inplace=True)
rma_w.sort('full_season',inplace=True)
vil_w.sort('full_season',inplace=True)
atm_w.sort('full_season',inplace=True)
```
